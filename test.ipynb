```python
import requests
import pandas
from sqlalchemy import create_engine , text
import psycopg2
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
url = ('https://newsapi.org/v2/everything?'
            'q=Apple&'
            'from=2026-06-10&to=2026-06-11&'
            'sortBy=popularity&'
            f'apiKey={os.getenv('NEWS_API_KEY')}')

response = requests.get(url)
print(response.status_code)
```

```python
data =response.json()
data
```

```python
articles = data['articles']
articles
```

```python
import pandas as pd
articles_df=pd.DataFrame(articles)
articles_df.head()
```

```python
articles_df['content']
```

```python
articles_df=articles_df.drop(columns=['source','urlToImage'])
articles_df.head()
```

```python
articles_df.rename(columns={'publishedAt':'published_at'}, inplace=True)
articles_df.head()
```

```python
DATABASE_NAME = os.getenv('DATABASE_NAME')
DATABASE_USER = os.getenv('DATABASE_USER')
DATABASE_PORT = os.getenv('DATABASE_PORT')
DATABASE_PASSWORD = os.getenv('DATABASE_PASSWORD')
DATABASE_HOST = os.getenv('DATABASE_HOST')

engine = create_engine(f'postgresql+psycopg2://{DATABASE_USER}:{DATABASE_PASSWORD}@{DATABASE_HOST}:{DATABASE_PORT}/{DATABASE_NAME}')
```

```python
with engine.connect() as conn:
    resort = conn.execute(text('select * from articles;'))
    for i in resort:
        print(i)
```

```python
articles_df.to_sql('articles',engine, if_exists='replace', index= False)
```
